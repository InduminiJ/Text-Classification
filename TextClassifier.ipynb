```
from sklearn.feature_extraction.text import CountVectorizer

def countVector_split(raw_data):
    vectorizer = CountVectorizer(binary=True, max_features=5000)
    X = vectorizer.fit_transform(raw_data.body) 

    return train_test_split(X, raw_data['subreddit_ENC'])
# raw_data['CountVector'] = X.toarray().tolist()
```

```
from sklearn.feature_extraction.text import TfidfVectorizer

def TfidfVector_split(raw_data):
    tfidf_vectorizer = TfidfVectorizer(stop_words='english', ngram_range=(1, 1), binary=True, max_features=5000)
    X_tfidf = tfidf_vectorizer.fit_transform(raw_data.body)
    # raw_data['TfidfVector'] = X_tfidf.toarray().tolist()
    return train_test_split(X_tfidf, raw_data['subreddit_ENC'])
```

```

class BernoulliNaiveBayes():
    def __init__(self):
        self.classes = None
        # {class: _, Pr: []}
        self.probabilities = None
        self.class_count = None

    def fit(self, X, y):
        self.classes = np.unique(y)
        for c in self.classes:
            #self.probabilities = dict(zip(self.classes, [{} for _ in range(len(self.classes))]))
            self.probabilities = dict(zip(self.classes, [[1]*len(X[0]) for _ in range(len(self.classes))]))
        self.class_count = dict(zip(self.classes, [2]*len(self.classes)))
        for i in range(len(y)):
            cl =  y[i]
            self.class_count[cl]+=1
            for j in range(len(X[i])):
                self.probabilities[cl][j] += X[i][j]
        for i in range(len(self.classes)):
            cl =  self.classes[i]
            self.probabilities[cl] = np.array(list(self.probabilities[cl]))/self.class_count[cl]
        print(self.probabilities)
        
    def predict(self, X):
        maxPr = 0
        bestCl = self.classes[0]
        for cl in self.classes:
            pr = self.class_count[cl]/sum(list(self.class_count.values()))
            for j in X:
                pr *= self.probabilities[cl][j] if X[j]==1 else 1 - self.probabilities[cl][j]
                
            # print(cl)
            # print(pr)
            if pr>maxPr:
                maxPr = pr
                bestCl = cl
        return bestCl

    def predict_arr(self, X):
        out = [-1]*len(X)
        for i in range(len(out)):
            out[i] = self.predict(X[i])
        return out

```
